```python
import os
import pandas as pd

# Directory containing the CSV files
directory = 'GroupData/'

# List to store individual DataFrames
data_frames = []

# Loop through the files in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        file_path = os.path.join(directory, filename)
        # Read each CSV file and append to the list
        data = pd.read_csv(file_path)
        data_frames.append(data)

# Concatenate all DataFrames in the list
combined_data = pd.concat(data_frames, ignore_index=True)

# Write the combined data to a single CSV file
combined_data.to_csv('CombinedData.csv', index=False)
```

```python
#drop all rows taht contains no data
filteredDf = df[df["extra_notes"] != "No Data"]
filteredDf = filteredDf[filteredDf["0w"].notnull()]
filteredDf = filteredDf[filteredDf["12w"].notnull()]
filteredDf = filteredDf[filteredDf["0w"]!=0]
len(filteredDf)
```

```python
contains2yearData = filteredDf["96w"].isnull().sum()
print(len(filteredDf) - contains2yearData)
print(contains2yearData/len(filteredDf))

contains1yearData = filteredDf["48w"].isnull().sum()
print(len(filteredDf) - contains1yearData)
print(contains1yearData/len(filteredDf))

print(len(filteredDf))

```

```python
cols = ["-8w%","-4w%","-2w%","-1w%"]
filteredDf["-8w%"] = 0
filteredDf["-4w%"] = 0
filteredDf["-2w%"] = 0
filteredDf["-1w%"] = 0
len(filteredDf)
```

```python
def getPercentChange():
    for i, row in filteredDf.iterrows():
        price = row["0w"]
        twoMonthPriorPrice = row["-8w"] if row["-8w"] else price
        oneMonthPriorPrice = row["-4w"] if row["-4w"] else price
        twoWeekPriorPrice = row["-2w"] if row["-2w"] else price 
        oneWeekPriorPrice = row["-1w"] if row["-1w"]else price
        threeMonthAfterPrice = row["26w"] #this will be our label

        if(twoMonthPriorPrice != 0):
            filteredDf.at[i, "-8w%"] = ((price-twoMonthPriorPrice) / twoMonthPriorPrice) * 100
        if(oneMonthPriorPrice != 0):
            filteredDf.at[i, "-4w%"] = ((price-oneMonthPriorPrice) / oneMonthPriorPrice) * 100
        if(twoWeekPriorPrice != 0):
            filteredDf.at[i, "-2w%"] = ((price-twoWeekPriorPrice) / twoWeekPriorPrice) * 100
        if(oneWeekPriorPrice != 0):
            filteredDf.at[i, "-1w%"] = ((price-oneWeekPriorPrice) / oneWeekPriorPrice) * 100

        if(threeMonthAfterPrice != 0):
            filteredDf.at[i, "26w%"] = ((threeMonthAfterPrice-price) / price) * 100

        

getPercentChange()
        
```

```python
filteredDf[cols] = filteredDf[cols].fillna(0)
filteredDf.tail(15)
```

```python
filteredDf['Type'] = filteredDf['Type'].fillna(-1)
filteredDf['Type'] = filteredDf['Type'].apply(lambda x: 1 if 'D' in str(x) else 0)
filteredDf.tail(10)
```
